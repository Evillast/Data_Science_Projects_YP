### Структура данных

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

- **`airports_nearest`** — расстояние до ближайшего аэропорта (м);
- **`balcony`** — число балконов;
- **`ceiling_height`** — высота потолков (м);
- **`cityCenters_nearest`** — расстояние до центра города (м);
- **`days_exposition`** — сколько дней было размещено объявление (от публикации до снятия);
- **`first_day_exposition`** — дата публикации;
- **`floor`** — этаж;
- **`floors_total`** — всего этажей в доме;
- **`is_apartment`** — апартаменты (булев тип);
- **`kitchen_area`** — площадь кухни (м²);
- **`last_price`** — цена на момент снятия с публикации;
- **`living_area`** — жилая площадь (м²);
- **`locality_name`** — название населённого пункта;
- **`open_plan`** — свободная планировка (булев тип);
- **`parks_around3000`** — число парков в радиусе 3 км;
- **`parks_nearest`** — расстояние до ближайшего парка (м);
- **`ponds_around3000`** — число водоёмов в радиусе 3 км;
- **`ponds_nearest`** — расстояние до ближайшего водоёма (м);
- **`rooms`** — число комнат;
- **`studio`** — квартира-студия (булев тип);
- **`total_area`** — общая площадь квартиры (м²);
- **`total_images`** — число фотографий квартиры в объявлении.

Пропуски в столбце с высотой потолков **'ceiling_height'** есть смысл заполнить медианными значениями. Однако в центре города, в старом фонде, высота потолков должна быть заметно выше, чем в новых районах города. Прэтому разобьем наши данные на группы в зависимости от удаленности от центра и будем считать медианы внутри каждой группы.

Предлагаем разбить все объекты на следующие категории: 
    
- **`А`** — 0 - 2500 м от центра;
- **`B`** — 2501 - 5000 м от центра;
- **`С`** — 5001 - 16000 м от центра;
- **`D`** — больше 16000 м от центра.

Протяжённость Санкт-Петербурга с севера на юг в пределах КАД — 32000 м, поэтому все, что находится дальше 16000 м от центра можно считать условным пригородом.

Напишем функцию, которая будет присваивать объекту категорию по удаленности от центра города:

```python
def categorize_center(distance):
    if distance <= 2500:
        return 'A'
    elif distance <= 5000:
        return 'B'
    elif distance <= 16000:
        return 'C'
    elif distance > 16000:
        return 'D'
```

Применим ее к датафрейму и создадим новый столбец **'center_category'**:

```python
data['center_category'] = data['cityCenters_nearest'].apply(categorize_center)
data[['ceiling_height','locality_name', 'cityCenters_nearest', 'center_category']].head(15)
```

Посчитаем медиану высоты для каждой категории (средним заполнять не будем, помня про выбросы — где-то встречались 100-метровые потолки):

```python
data.groupby('center_category')['ceiling_height'].median().to_frame()
```

Медианы получились действительно разными. Теперь заполним ими пропуски:

```python
cat_list = ['A', 'B', 'C', 'D']
for cat in cat_list:
    median_height = data.groupby('center_category')['ceiling_height'].median().get(key=cat)
    data['ceiling_height'] = data.groupby('center_category')['ceiling_height'].apply(lambda x: x.fillna(x.median()))
```

Предлагаем найти средние отношения жилой площади к общей площади и площади кухни к общей площади, а затем, зная эти сооотношения, вычислить жилую площадь и площадь кухни для тех строк, где она не указана.

Видим, что **сильнее всего на стоимость квартиры влияют общая площадь, жилая площадь и площадь кухни, а также количество комнат и высота потолков.** Также влияние оказывают **удаленность от центра Санкт-Петербурга** (чем дальше от центра, тем ниже стоимость) и **год публикации объявления**. Небольшое влияние оказывает присутствие парков и водоемов поблизости, а также свободная планировка. Причем многие из этих параметров очень сильно коррелируют между собой (так, например, жилая площадь и площадь кухни сильно коррелируют с общей площадью, количество комнат сильно коррелирует с общей площадью и жилой площадью, а высота потолков, как мы помним, коррелирует с удаленностью от центра — чем дальше от центра, тем ниже потолки).

```python
print('Коэффициент корреляции жилой и общей площади:', '{:.2}'.format(data['living_area'].corr(data['total_area'])))
print('Коэффициент корреляции площади кухни и общей площади:', '{:.2}'.format(data['kitchen_area'].corr(data['total_area'])))
print('Коэффициент корреляции количества комнат и общей площади:', '{:.2}'.format(data['rooms'].corr(data['total_area'])))
print('Коэффициент корреляции количества комнат и жилой площади:', '{:.2}'.format(data['rooms'].corr(data['living_area'])))
print('Коэффициент корреляции высоты потолков и удаленности от центра:',
      '{:.2}'.format(data['ceiling_height'].corr(data['cityCenters_nearest'])))
```
