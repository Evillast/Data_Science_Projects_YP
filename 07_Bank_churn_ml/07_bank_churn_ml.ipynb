## Введение

Из «Бета-Банка» стали уходить клиенты.  Каждый месяц. Немного, но заметно. Маркетологи посчитали: сохранять текущих клиентов дешевле, чем привлекать новых.

### Цели и задачи

**Цель исследования:**

Построить модель машинного обучения, которая сможет спрогнозировать, уйдёт ли клиент из банка в ближайшее время или нет, на основании исторических данных о поведении клиентов и расторжении договоров с банком.

**Задачи:**

1. Изучить данные и подготовить их для дальнейшей работы;
2. Исследовать баланс классов и устранить дисбаланс при необходимости;
3. Построить модель машинного обучения с предельно большим значением *F1-меры* (не менее 0.59);
4. Измерить *AUC-ROC*, сравнить её значение с *F1-мерой*.

### Структура данных

**Признаки:**

- **`RowNumber`** — индекс строки в данных;
- **`CustomerId`** — уникальный идентификатор клиента;
- **`Surname`** — фамилия;
- **`CreditScore`** — кредитный рейтинг;
- **`Geography`** — страна проживания;
- **`Gender`** — пол;
- **`Age`** — возраст;
- **`Tenure`** — сколько лет человек является клиентом банка;
- **`Balance`** — баланс на счёте;
- **`NumOfProducts`** — количество продуктов банка, используемых клиентом;
- **`HasCrCard`** — наличие кредитной карты;
- **`IsActiveMember`** — активность клиента;
- **`EstimatedSalary`** — предполагаемая зарплата;

**Целевой признак:**

- **`Exited`** — факт ухода клиента.

### Предобработка данных

Названия столбцов прописаны *CamelCase*-ом. Для удобства дальнейшей работы переведем их в *snake_case*:

```python
data.columns = data.columns.str.lower()
data = data.rename(columns={'rownumber':'row_number', 
                        'customerid':'customer_id',  
                        'creditscore':'credit_score', 
                        'numofproducts':'num_of_products', 
                        'hascrcard':'has_cr_card',
                        'isactivemember':'is_active_member', 
                        'estimatedsalary':'estimated_salary'
                           })
```

Проверим:

```python
data.columns.tolist()
```

Удалим столбцы, бесполезные для обучения модели:

```python
data = data.drop(['row_number', 'customer_id', 'surname'], axis=1)
```

```python
data.head()
```

Для дальнейшей работы необходимо разделить данные на выборки. Разобьем данные на три выборки: обучающую, валидационную и тестовую в соотношении 3-1-1:

```python
features = data.drop('exited', axis=1)
target = data['exited']
```

```python
# Делаем валидационную выборку со стратификацией по целевому признаку
features_train, features_valid, target_train, target_valid = train_test_split(
    features, target, test_size=0.2, random_state=12345, stratify=target)
```

```python
# Повторяем для тестовой выборки, но с test_size=0.25, т.к. 0.25*(1-0.2)=0.2
features_train, features_test, target_train, target_test = train_test_split(
    features_train, target_train, test_size=0.25, random_state=12345, stratify=target_train)
```

```python
print(f'Размеры обучающей выборки: признаки: {features_train.shape}, целевой признак: {target_train.shape}')
print(f'Размеры валидационной выборки: признаки: {features_valid.shape}, целевой признак: {target_valid.shape}')
print(f'Размеры тестовой выборки: признаки: {features_test.shape}, целевой признак: {target_test.shape}')
```

Воспользуемся техникой **One-Hot Encoding**, чтобы преобразовать категориальные признаки в числовые.

**ОHE** переводит категориальные признаки в числовые в два этапа:
- Для каждого значения признака создаётся новый столбец;
- Если объекту категория подходит, присваивается значение '1', если нет — '0'.

Таким образом создаются *фиктивные переменные* (*dummy variables*).

```python
categorical = ['geography', 'gender']
```

```python
encoder = OneHotEncoder(drop='first', sparse=False)

features_train_ohe = pd.DataFrame(encoder.fit_transform(features_train[categorical]),
                                                        columns=['geography_1', 'geography_2', 'gender_1'])
features_valid_ohe = pd.DataFrame(encoder.transform(features_valid[categorical]),
                                                        columns=['geography_1', 'geography_2', 'gender_1'])
features_test_ohe = pd.DataFrame(encoder.transform(features_test[categorical]),
                                                        columns=['geography_1', 'geography_2', 'gender_1'])
```

Удалим из выборок столбцы, которые мы собираемся заменить:

```python
features_train = features_train.drop(['geography', 'gender'], axis=1).reset_index(drop=True)
features_valid = features_valid.drop(['geography', 'gender'], axis=1).reset_index(drop=True)
features_test = features_test.drop(['geography', 'gender'], axis=1).reset_index(drop=True)
```

А теперь добавим в выборки преобразованные столбцы:

```python
features_train = pd.concat([features_train, features_train_ohe], axis=1)
features_valid = pd.concat([features_valid, features_valid_ohe], axis=1)
features_test = pd.concat([features_test, features_test_ohe], axis=1)
```

Посмотрим, что получилось:

```python
features_train.head()
```

```python
features_valid.head()
```

```python
features_test.head()
```

Теперь разберемся с пропусками в столбце **'tenure'**. Удалять строки с пропусками не вариант — их слишком много.

Из отчета, приведенного выше, видно, что этот столбец практически не коррелирует с остальными данными в таблице.

Заполним пропуски. Для этого используем **KNNIImputer**. Он производит вменение для заполнения пропущенных значений с использованием k-ближайших соседей. Пропущенные значения каждой выборки вменяются с использованием среднего значения ближайших соседей, найденных в обучающем наборе. Две выборки близки, если близки признаки, которые отсутствуют в каждой из них.

```python
imputer = KNNImputer(missing_values=np.nan, n_neighbors=10, weights='distance')

features_train = pd.DataFrame(imputer.fit_transform(features_train).round(), columns=features_train.columns)
features_valid = pd.DataFrame(imputer.transform(features_valid).round(), columns=features_valid.columns)
features_test = pd.DataFrame(imputer.transform(features_test).round(), columns=features_test.columns)
```

Проверим количество пропусков:

```python
print('Количество пропусков в столбце "tenure"')
print(f'в тренировочной выборке: {features_train["tenure"].isna().sum()}')
print(f'в валидационной выборке: {features_valid["tenure"].isna().sum()}')
print(f'в тестовой выборке: {features_test["tenure"].isna().sum()}')
```

Из-за того, что мы переводили **features** в numpy.ndarray и обратно в датафрэймы, там изменилась индексация. Значит, нужно поправить индексацию и в **target**.

```python
target_train.reset_index(drop=True, inplace=True)
target_valid.reset_index(drop=True, inplace=True)
target_test.reset_index(drop=True, inplace=True)
```

И, наконец, стандартизуем числовые признаки для лучшей интерпретируемости линейной модели с помощью **StandardScaler**:

```python
features_train_st = features_train.copy()
features_valid_st = features_valid.copy()
features_test_st = features_test.copy()
```

```python
pd.options.mode.chained_assignment = None
```

```python
numeric = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products', 'estimated_salary']
```

```python
scaler = StandardScaler()

features_train_st[numeric] = scaler.fit_transform(features_train[numeric])
features_valid_st[numeric] = scaler.transform(features_valid[numeric])
features_test_st[numeric] = scaler.transform(features_test[numeric])
```

### Обзор данных (продолжение)

**F1-мерa** — среднее гармоническое метрик **recall** (полнота) и **precision** (точность).

**AUC-ROC** — площадь под кривой ошибок (***ROC-curve***, отражает изменение соотношения ложноположительных ответов *(FPR)* и истинно положительных ответов *(TPR)* при изменении порога классификации).

Для удобства дальнейшей работы создадим объекты *Series*, куда будем сохранять лучше результаты *F1-меры* для каждой модели:

```python
unbalanced_score = pd.Series({'unbalanced':[]}, index=[
    'decision_tree', 'random_forest', 'adaptive_boosting', 'gradient_boosting', 'logistic_regression'])
```

А также напишем функцию для построения графика *ROC-curve* и расчета *AUC-ROC*:

```python
def roc_curve_plot(model, model_name, f, t):
    
    probabilities = model.predict_proba(f)
    probabilities_one = probabilities[:, 1]
    
    fpr, tpr, thresholds = roc_curve(t, probabilities_one)
    plt.figure(figsize=(5,5))
    plt.title('ROC-кривая модели {}'.format(model_name), fontsize=14, pad=20)
    plt.plot([0, 1], [0, 1], color='royalblue', linestyle='--')
    plt.plot(fpr, tpr, color='orange')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate', fontsize=12)
    plt.ylabel('True Positive Rate', fontsize=12)
    plt.show()
    
    print('AUC-ROC модели: {:.3f}'.format(roc_auc_score(t, probabilities_one)))
```

### Дерево решений

**Дерево решений** — универсальный алгоритм машинного обучения. Структура дерева представляет собой «листья» и «ветки». На рёбрах («ветках») дерева решения записаны признаки, от которых зависит целевая переменная, а в «листьях» записаны значения целевой переменной. Чтобы классифицировать новый случай, надо спуститься по дереву до листа и выдать соответствующее значение.

В дальнейшем, когда мы будем исследовать методы борьбы с дисбалансом, нам еще не раз предстоит обращаться к этой и остальным моделям. Поэтому превратим их в функции.

```python
def gparms_tree(f_t, t_t, balanced, threshold, table):
    best_depth = 0
    best_threshold = 0
    best_result = 0
    best_model = None
    if threshold == False:
        for depth in tqdm(range(2, 16)):
            if balanced == False:
                model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
            elif balanced:
                model = DecisionTreeClassifier(random_state=12345, max_depth=depth, class_weight='balanced')
            model.fit(f_t, t_t)
            predictions_valid = model.predict(features_valid)
            result = f1_score(target_valid, predictions_valid)
            if result > best_result:
                best_depth = depth
                best_result = result
                best_model = model
        table.loc['decision_tree'] = best_result.round(3)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    elif threshold:
        for depth in tqdm(range(2, 16)):
            model = DecisionTreeClassifier(random_state=12345, max_depth=depth)
            model.fit(f_t, t_t)
            probabilities_valid = model.predict_proba(features_valid)
            probabilities_one_valid = probabilities_valid[:, 1]
            for threshold in np.arange(0.00, 1.02, 0.02):
                predictions_valid = probabilities_one_valid > threshold
                result = f1_score(target_valid, predictions_valid)
                if result > best_result:
                    best_depth = depth
                    best_threshold = threshold
                    best_result = result
                    best_model = model
        table.loc['decision_tree'] = best_result.round(3)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('Лучший порог классификации на валидационной выборке:', best_threshold)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    return best_model
```

### Случайный лес

**Случайный лес** —  алгоритм машинного обучения, использующий большой ансамбль независимых друг от друга решающих деревьев и принимающий решение на основе голосования.

```python
def gparms_forest(f_t, t_t, balanced, threshold, table):
    best_estimators = 0
    best_depth = 0
    best_threshold = 0
    best_result = 0
    best_model = None
    if threshold == False:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                if balanced == False:
                    model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
                elif balanced:
                    model = RandomForestClassifier(random_state=12345, n_estimators=est,
                                                   max_depth=depth, class_weight='balanced')
                model.fit(f_t, t_t)
                predictions_valid = model.predict(features_valid)
                result = f1_score(target_valid, predictions_valid)
                if result > best_result:
                    best_estimators = est
                    best_depth = depth
                    best_result = result
                    best_model = model
        table.loc['random_forest'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    elif threshold:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
                model.fit(f_t, t_t)
                probabilities_valid = model.predict_proba(features_valid)
                probabilities_one_valid = probabilities_valid[:, 1]
                for threshold in np.arange(0.00, 1.02, 0.02):
                    predictions_valid = probabilities_one_valid > threshold
                    result = f1_score(target_valid, predictions_valid)
                    if result > best_result:
                        best_estimators = est
                        best_depth = depth
                        best_threshold = threshold
                        best_result = result
                        best_model = model
        table.loc['random_forest'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('Лучший порог классификации на валидационной выборке:', best_threshold)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    return best_model
```

### Адаптивный бустинг

**Адаптивный бустинг** — алгоритм машинного обучения, который создает решающую модель прогнозирования в виде ансамбля слабых моделей прогнозирования, обычно деревьев решений. Является адаптивным в том смысле, что последующие слабые модели настраиваются в пользу тех экземпляров, которые были неправильно классифицированы предыдущими классификаторами. Затем слабые модели объединяются во взвешенную сумму, которая представляет окончательный результат усиленного классификатора.

```python
def gparms_adaboost(f_t, t_t, balanced, threshold, table):
    best_estimators = 0
    best_depth = 0
    best_threshold = 0
    best_result = 0
    best_model = None
    if threshold == False:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                if balanced == False:
                    base = DecisionTreeClassifier(random_state=12345, max_depth=depth)
                    model = AdaBoostClassifier(random_state=12345, base_estimator=base, n_estimators=est)
                elif balanced:
                    base = DecisionTreeClassifier(random_state=12345, max_depth=depth, class_weight='balanced')
                    model = AdaBoostClassifier(random_state=12345, base_estimator=base, n_estimators=est)
                model.fit(f_t, t_t)
                predictions_valid = model.predict(features_valid)
                result = f1_score(target_valid, predictions_valid)
                if result > best_result:
                    best_estimators = est
                    best_depth = depth
                    best_result = result
                    best_model = model
        table.loc['adaptive_boosting'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина базового дерева:', best_depth)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    elif threshold:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
                model.fit(f_t, t_t)
                probabilities_valid = model.predict_proba(features_valid)
                probabilities_one_valid = probabilities_valid[:, 1]
                for threshold in np.arange(0.00, 1.02, 0.02):
                    predictions_valid = probabilities_one_valid > threshold
                    result = f1_score(target_valid, predictions_valid)
                    if result > best_result:
                        best_estimators = est
                        best_depth = depth
                        best_threshold = threshold
                        best_result = result
                        best_model = model
        table.loc['adaptive_boosting'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина базового дерева:', best_depth)
        print('Лучший порог классификации на валидационной выборке:', best_threshold)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    return best_model
```

### Градиентный бустинг

**Градиентный бустинг** — еще один алгоритм машинного обучения, который создает решающую модель прогнозирования в виде ансамбля слабых моделей прогнозирования, обычно также деревьев решений. При каждом следующем шаге новая модель обучается с использованием данных об ошибках предыдущих. Результирующая функция представляет собой линейную комбинацию базовых, слабых моделей.

Разница между градиентным и адаптивным бустингом заключается в том, как алгоритмы идентифицируют слабые модели. Адаптивный бустинг выявляет их на основании высоких значений весов, а градиентный бустинг – на основании градиентов функции потерь.

```python
def gparms_gradboost(f_t, t_t, threshold, table):
    best_estimators = 0
    best_depth = 0
    best_threshold = 0
    best_result = 0
    best_model = None
    if threshold == False:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                model = GradientBoostingClassifier(random_state=12345, n_estimators=est, max_depth=depth)
                model.fit(f_t, t_t)
                predictions_valid = model.predict(features_valid)
                result = f1_score(target_valid, predictions_valid)
                if result > best_result:
                    best_estimators = est
                    best_depth = depth
                    best_result = result
                    best_model = model
        table.loc['gradient_boosting'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    elif threshold:
        for est in tqdm(range(20, 101, 5)):
            for depth in range(2, 16):
                model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
                model.fit(f_t, t_t)
                probabilities_valid = model.predict_proba(features_valid)
                probabilities_one_valid = probabilities_valid[:, 1]
                for threshold in np.arange(0.00, 1.02, 0.02):
                    predictions_valid = probabilities_one_valid > threshold
                    result = f1_score(target_valid, predictions_valid)
                    if result > best_result:
                        best_estimators = est
                        best_depth = depth
                        best_threshold = threshold
                        best_result = result
                        best_model = model
        table.loc['gradient_boosting'] = best_result.round(3)
        print('Лучшее количество деревьев:', best_estimators)
        print('Лучшая максимально допустимая глубина дерева:', best_depth)
        print('Лучший порог классификации на валидационной выборке:', best_threshold)
        print('F1-мера лучшей модели на валидационной выборке:', '{:.3f}'.format(best_result))
    return best_model
```

### Логистическая регрессия

**Логистическая регрессия** — алгоритм машинного обучения, используемый для прогнозирования вероятности целевой переменной. Природа целевой переменной дихотомична. т.е. существует только два возможных класса. Логистическая регрессия вычисляет, к какому классу близок объект, и в зависимости от ответа выбирает нужный класс.

```python
def log_reg(f_t, t_t, balanced, threshold, table):
    best_threshold = 0
    best_result = 0
    if threshold == False:
        if balanced == False:
            model = LogisticRegression(random_state=12345, solver='lbfgs')
        elif balanced:
            model = LogisticRegression(random_state=12345, solver='liblinear', class_weight='balanced')
        model.fit(f_t, t_t)
        predictions_valid = model.predict(features_valid_st)
        result = f1_score(target_valid, predictions_valid)
        table.loc['logistic_regression'] = result.round(3)
        print('F1-мера модели на валидационной выборке:', '{:.3f}'.format(result))
    elif threshold:
        model = LogisticRegression(random_state=12345, solver='lbfgs')
        model.fit(f_t, t_t)
        probabilities_valid = model.predict_proba(features_valid_st)
        probabilities_one_valid = probabilities_valid[:, 1]
        for threshold in tqdm(np.arange(0.00, 1.02, 0.02)):
            predictions_valid = probabilities_one_valid > threshold
            result = f1_score(target_valid, predictions_valid)
            if result > best_result:
                best_threshold = threshold
                best_result = result
        table.loc['logistic_regression'] = best_result.round(3)
        print('Лучший порог классификации на валидационной выборке:', best_threshold)
        print('F1-мера модели на валидационной выборке:', '{:.3f}'.format(best_result))
    return model
```

### Увеличение выборки

**Увеличение выборки** — увеличение количества объектов миноритарного класса в обучающей выборке.

Напишем функцию, которая увеличит нашу обучающую выборку и перемешает ее:

```python
def upsample(f_t, f_st, t_t, repeat):
    features_zeros = f_t[t_t == 0]
    features_ones = f_t[t_t == 1]
    features_st_zeros = f_st[t_t == 0]
    features_st_ones = f_st[t_t == 1]
    target_zeros = t_t[t_t == 0]
    target_ones = t_t[t_t == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    features_st_upsampled = pd.concat([features_st_zeros] + [features_st_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, features_st_upsampled, target_upsampled = shuffle(
        features_upsampled, features_st_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, features_st_upsampled, target_upsampled
```

Увеличим количество уходящих клиентов в выборке в 4 раза, так как 20% клиентов уходят из банка, а 80% остаются — уходящих в 4 раза меньше. Также увеличим стандартизированную выборку для линейной модели.

### Уменьшение выборки

**Уменьшение выборки** — уменьшение количества объектов мажоритарного класса в обучающей выборке.

Напишем функцию, которая уменьшит обучающую выборку и перемешает ее:

```python
def downsample(f_t, f_st, t_t, fraction):
    features_zeros = f_t[t_t == 0]
    features_ones = f_t[t_t == 1]
    features_st_zeros = f_st[t_t == 0]
    features_st_ones = f_st[t_t == 1]
    target_zeros = t_t[t_t == 0]
    target_ones = t_t[t_t == 1]
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    features_st_downsampled = pd.concat(
        [features_st_zeros.sample(frac=fraction, random_state=12345)] + [features_st_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    features_downsampled, features_st_downsampled, target_downsampled = shuffle(
        features_downsampled, features_st_downsampled, target_downsampled, random_state=12345)
    return features_downsampled, features_st_downsampled, target_downsampled
```
