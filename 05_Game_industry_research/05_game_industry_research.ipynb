Заказчик — интернет-магазин «Стримчик», который продаёт компьютерные игры по всему миру. 

### Цели и задачи

**Цель исследования:**

Изучить доступные из открытых источников исторические данные о продажах игр, оценках пользователей и экспертов, жанрах и платформах и выявить определяющие успешность игры закономерности, чтобы предсказывать потенциальный успех продукта и планировать рекламные кампании.

**Задачи:**

1. Провести обзор данных;
2. Выполнить предобработку данных:
   - Выполнить преобразование типов данных в тех столбцах, где это необходимо;
   - Обработать пропуски там, где это возможно;
   - Проверить данные на наличие дубликатов.
3. Провести исследовательский анализ данных:
   - Изучить динамику выпуска игр за весь период;
   - Изучить динамику продаж по платформам по годам. Определить средний срок жизни платформы;
   - Выбрать актуальный период для анализа и изучить:
     - Динамику продаж по платформам. Определить, какие платформы лидируют по продажам, растут и падают. Выбрать несколько потенциально прибыльных платформ;
     - Изучить, как влияют на продажи отзывы пользователей и критиков;
     - Изучить общее распределение игр по жанрам. Найти жанры с самыми высокими и низкими продажами.
     - Составить портрет пользователя каждого региона (Северная Америка, Европа, Япония):
       - Определить самые популярные платформы;
       - Самые популярные жанры;
       - Определить, влияет ли рейтинг ESRB на продажи в регионе.
4. Проверить следующие гипотезы:
   - Средние пользовательские рейтинги платформ Xbox One и PC одинаковы;

### Структура данных

- **`Name`** — название игры;
- **`Platform`** — платформа;
- **`Year_of_Release`** — год выпуска;
- **`Genre`** — жанр игры;
- **`NA_sales`** — продажи в Северной Америке (млн копий);
- **`EU_sales`** — продажи в Европе (млн копий);
- **`JP_sales`** — продажи в Японии (млн копий);
- **`Other_sales`** — продажи в других странах (млн копий);
- **`Critic_Score`** — оценка критиков (макс. 100);
- **`User_Score`** — оценка пользователей (макс. 10);
- **`Rating`** — возрастной рейтинг от ESRB.

## Предобработка данных

### Изменение регистра названий столбцов

Для удобства приведем названия столбцов к нижнему регистру:

```python
data.columns = data.columns.str.lower()
```

Проверим:

```python
data.columns.tolist()
```

### Обработка пропусков

Посмотрим на столбцы, в которых содержатся пропущенные значения:

```python
pd.DataFrame(data.isna().sum().sort_values(ascending=False), columns=['NaN']).style.background_gradient('coolwarm')
```

Видим, что пропуски содержатся в 6 столбцах: **'critic_score'**, **'rating'**, **'user_score'**, **'year_of_release'**, **'name'**, **'genre'**. Рассмотрим их подробнее.

Для начала проверим, не связаны ли между собой пропуски в столбцах **'critic_score'**, **'user_score'** и **'rating'**:

```python
print('Строк с пропусками одновременно в 3 столбцах:',
      data.query('critic_score.isna() and user_score.isna() and rating.isna()').shape[0])
```

Очевидно, что раз большая часть пропусков появляется одновременно в трех столбцах, эти пропуски как-то взаимосвязаны между собой. Теперь проверим, не связаны ли они с годом выпуска игр — возможно, слишком старым играм такие оценки не присваивались, а слишком новые еще не успели оценить.

```python
data.query('critic_score.isna() or user_score.isna() or rating.isna()')['year_of_release'].value_counts().to_frame()
```

Видим, что пропуски встречаются практически в каждом году. Похоже, что это неслучайные пропуски, вызванные проблемами при выгрузке данных. И заполнить эти пропуски какими-либо усредненными данными ни по одному столбцу не представляется возможным.

Теперь посмотрим внимательнее на столбец **'user_score'**. Судя по информации, которую мы получили методом info(), этот столбец имеет тип 'object', хотя должен содержать только числа.

```python
data['user_score'].unique()
```

Среди значений нам встретилась аббревиатура **«tbd»**.

**«TBD»** — аббревиатура от английского **«To Be Determined»** (*«будет определено»*) или **«To Be Decided»** (*«будет решено»*). Используется, если какая-то информация еще не определена или решение по вопросу не принято. 

Так в разделе FAQ на популярном ресурсе https://www.metacritic.com/ написано, что эта аббревиатура используется тогда, когда игра имеет недостаточное количество отзывов (менее 4) для подсчета рейтинга пользователей и его публикации.

Такое возможно с малопопулярными играми или играми, имеющими в основном неанглоязычную аудиторию — например, японскую.

Так как аббревиатура **«tbd»** фактически является таким же пропуском, заменим ее на обычное значение *NaN*, чтобы иметь возможность привести столбец к числовому типу и дальше анализировать его.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = np.nan
```

Посмотрим, что все сработало корректно:

```python
print('Значений "tbd" в столбце "user_score":', data.query('user_score == "tbd"')['user_score'].count())
print('Пропусков в столбце "user_score":', data['user_score'].isna().sum())
```

А вот в столбце **'rating'** все значения действительно текстовые, т.к. рейтинг ESRB обозначается буквами. Таким образом мы можем заменить пропуски в этом столбце на значение 'unknown'.

```python
data['rating'] = data['rating'].fillna('unknown')
```

Проверим:

```python
print('Пропусков в столбце "rating":', data['rating'].isna().sum())
```

Двигаемся дальше. В столбце **'year_of_release'** пропусков, к счастью, не так много. Посмотрим, связаны ли эти пропуски с каким-либо другими:

```python
print('Строк с парными пропусками:',
      data.query(('critic_score.isna() or user_score.isna()') and 'year_of_release.isna()').shape[0])
```

Видим, что абсолютно все строки, в которых пропущено значение в столбце **'year_of_release'**, имеют пропуски в столбцах **'critic_score'** или **'user_score'** (на столбец **'rating'** не смотрим, так как уже заполнили в нем пропуски). Возможно, те же самые проблемы с выгрузкой, тогда пропуски в столбце **'year_of_release'** являются неслучайными. В то же время пропусков в столбце с годами несопоставимо меньше. 

Мы не видим возможности как-либо заполнить эти пропуски. Т.к. для нашего анализа год выпуска игры очень важен, а строк с пропусками в столбце **'year_of_release'** не так уж и много, просто удалим такие строки.

```python
data = data.dropna(subset=['year_of_release']).reset_index(drop=True)
```

Проверим:

```python
print('Пропусков в столбце "year_of_release":', data['year_of_release'].isna().sum())
```

У нас осталось по 2 пропуска в столбцах **'name'** и  **'genre'**. Проверим, связаны ли они друг с другом:

```python
print('Строк с парными пропусками:',
      data.query('name.isna() and genre.isna()').shape[0])
```

Таким образом эти пропуски тоже можно считать неслучайными. Заполнить их невозможно, но это не имеет никакого значения, учитывая их количество. Поэтому просто удалим их:

```python
data = data.dropna(subset=['name']).reset_index(drop=True)
```

Еще раз посмотрим на количество пропусков в таблице:

```python
pd.DataFrame(data.isna().sum().sort_values(ascending=False), columns=['NaN']).style.background_gradient('coolwarm')
```

**Вывод:**

Мы обнаружили пропуски в 6 столбцах датафрэйма из 11: **'critic_score'**, **'rating'**, **'user_score'**, **'year_of_release'**, **'name'**, **'genre'**. Заполнить их смогли только в столбце **'rating'** — но не как-либо качественно, а просто заменили на значение *'unknown'*. Строки с пропусками в столбцах **'year_of_release'**, **'name'** и **'genre'** мы просто удалили, т.к. их немного и эта информация первостепенно важна для исследования. Также мы обнаружили аббревиатуру *«tbd»* в столбце **'user_score'** и выяснили, что она фактически является тем же пропуском, поэтому заменили ее на обычные значения *NaN*.

### Распределение игр по жанрам. Влияние жанра игры на продажи

Теперь исследуем игровые жанры и их влияние на продажи игр.

Для начала построим сводную таблицу по жанрам игр в актуальный период времени:

```python
# Строим сводную таблицу с группировкой по жанру
# Cчитаем общее количество игр в жанре и общее количество продаж по жанру
pivot_genre = data_act.pivot_table(aggfunc={'name': 'count', 'total_sales': 'sum'},
                                   index='genre').sort_values(by='name', ascending=False)

# Для наглядности переименуем столбец 'name' в 'total_games_amount'
pivot_genre.rename(columns={'name': 'total_games_amount'}, inplace=True)

# Добавим столбцы со средним количеством выпускаемых игр в год
# и средним количеством продаж в год для каждого жанра
# Делим на 5, поскольку рассматриваем актуальный период с 2012 по 2016 год
pivot_genre['games_per_year'] = (pivot_genre['total_games_amount'] / 5).astype(int)
pivot_genre['sales_per_year'] = (pivot_genre['total_sales'] / 5).round(2)

# Добавим столбец со средним количеством продаж на одну игру для каждого жанра
pivot_genre['sales_per_game'] = (pivot_genre['total_sales'] / pivot_genre['total_games_amount']).round(2)

pivot_genre
```

Сделать какие-либо выводы по такой таблице довольно сложно. Выделим топ-5 самых популярных платформ для каждого региона и представим их на графиках в процентных долях продаж:

```python
# Создаем словарь регионов - он поможет сделать понятные заголовки для графиков
dict_of_regs = {'eu_sales': 'Европа', 'jp_sales': 'Япония',
    'na_sales': 'Северная Америка', 'other_sales': 'Другие регионы'}

# Цветовая палитра
colors = ['lightcoral', 'royalblue', 'plum', 'cyan',  'palegreen', 'silver']
```

## Статистический анализ

### Проверка гипотезы о равенстве средних пользовательских рейтингов платформ Xbox One и PC

Сформулируем нулевую и альтернативную гипотезы. За нулевую гипотезу традиционно возьмем гипотезу о равенстве:

- **Н₀ (нулевая гипотеза):** средние пользовательские рейтинги платформ **Xbox One** и **PC** одинаковые.

- **Н₁ (альтернативная гипотеза):** средние пользовательские рейтинги платформ **Xbox One** и **PC** отличаются.

Т.к. игры на **PC** выходят уже много лет, а платформа **Xbox One** вышла только в 2013 году, для объективности будем исследовать данные начиная с 2013 года.

Наша нулевая гипотеза —  гипотеза о равенстве средних двух генеральных совокупностей. Применим двусторонний тест, так как нам сейчас неважно, у какой платформы больше средний рейтинг, если они отличаются:

```python
alpha = .05
```

```python
xone = data_act.query('platform == "XOne" and not user_score.isna()')['user_score']
pc = data_act.query('platform == "PC" and year_of_release >= 2013 and not user_score.isna()')['user_score']
```

```python
results = st.ttest_ind(xone, pc)

print('p-value:', '{:.1f}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Видим, что на имеющихся данных на уровне статистической значимости 5% отвергнуть нашу гипотезу не удалось. Проверим вручную:

```python
print('Средний пользовательский рейтинг платформы "XOne":', '{:.1f}'.format(xone.mean()))
print('Средний пользовательский рейтинг платформы "PC":', '{:.1f}'.format(pc.mean()))
```

На самом деле рейтинги несколько отличаются, хотя и не слишком существенно.

### Проверка гипотезы о неравенстве средних пользовательских рейтингов  жанров Action и Sports

Сформулируем нулевую и альтернативную гипотезы. За нулевую гипотезу снова возьмем гипотезу о равенстве:

- **Н₀ (нулевая гипотеза):** средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.

- **Н₁ (альтернативная гипотеза):** средние пользовательские рейтинги жанров **Action** и **Sports** отличаются.

Будем исследовать данные за актуальный период — начиная с 2012 года.

Наша нулевая гипотеза — гипотеза о равенстве средних двух генеральных совокупностей. Применим двусторонний тест, так как нам сейчас неважно, у какого жанра больше средний рейтинг, если они отличаются:

```python
action = data_act.query('genre == "Action" and not user_score.isna()')['user_score']
sports = data_act.query('genre == "Sports" and not user_score.isna()')['user_score']
```

```python
results = st.ttest_ind(action, sports)

print('p-value:', '{:.1}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом на имеющихся данных на уровне статистической значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений пользовательских рейтингов жанров **Action** и **Sports** в пользу альтернативной гипотезы. Значит, наша изначальная гипотеза верна: средние пользовательские рейтинги данных жанров отличаются. Проверим, в чью пользу. Сформулируем новую альтернативную гипотезу:

- **Н₀ (нулевая гипотеза):** средние пользовательские рейтинги жанров **Action** и **Sports** одинаковые.

- **Н₁ₐ (новая альтернативная гипотеза):** средний пользовательские рейтинг жанра **Action** выше, чем средний пользовательский рейтинг жанра **Sports**.

Применим односторонний тест для рейтинга жанра **Action**:

```python
print('p-value:', '{:.1}'.format(results.pvalue / 2))

if (results.pvalue / 2 < alpha) and (action.mean() > sports.mean()):
    print('Отвергаем нулевую гипотезу: средний пользовательский рейтинг жанра Action выше среднего рейтинга жанра Sports')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Проверим вручную:

```python
print('Средний пользовательский рейтинг жанра "Action":', '{:.1f}'.format(action.mean()))
print('Средний пользовательский рейтинг жанра "Sports":', '{:.1f}'.format(sports.mean()))
```

Наша новая альтернативная гипотеза оказалась верной: средний пользовательский рейтинг жанра **Action** оказался выше среднего пользовательского рейтинга жанра **Sports**.
