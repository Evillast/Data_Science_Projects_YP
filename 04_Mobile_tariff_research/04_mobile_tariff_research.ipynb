## Введение

Компания-оператор мобильной связи «Мегалайн» предлагает клиентам два тарифных плана на выбор: **«Смарт»** и **«Ультра»**. Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.

В нашем распоряжении данные 500 пользователей «Мегалайна»: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018-й год. 

### Цели и задачи

**Цель исследования:**

Проанализировать поведение клиентов и сделать вывод: какой тариф, — **«Смарт»** или **«Ультра»**, — приносит компании больше денег.

**Задачи:** 
1. Провести обзор данных;
2. Провести предобработку данных, если это необходимо; 
3. Посчитать для каждого пользователя по месяцам:
   - количество сделанных звонков;
   - количество израсходованных минут разговора;
   - количество отправленных сообщений; 
   - количество потраченных мегабайт интернет-трафика;
4. Провести исследовательский анализ данных и посчитать помесячную выручку с каждого пользователя;
5. Проверить гипотезы:
   - Средняя выручка с пользователей тарифов «Ультра» и «Смарт» различается;
   - Средняя выручка с пользователей из Москвы отличается от выручки с пользователей других регионов.

### Описание тарифов

**Тариф «Смарт»**
- Ежемесячная плата: 550 рублей;
- Включено: 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика;
- Стоимость услуг сверх тарифного пакета:
  - Минута разговора — 3 рубля;
  - 1 сообщение — 3 рубля;
  - 1 Гб интернет-трафика — 200 рублей.
  
**Тариф «Ультра»**
- Ежемесячная плата: 1950 рублей;
- Включено: 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика;
- Стоимость услуг сверх тарифного пакета:
  - Минута разговора — 1 рубль;
  - 1 сообщение — 1 рубль;
  - 1 Гб интернет-трафика: 150 рублей.
  
Количество использованных минут и мегабайтов «Мегалайн» всегда округляет вверх. Если пользователь проговорил всего 1 секунду, в тарифе засчитывается целая минута.

### Структура данных

Данные находятся в 5 таблицах.

Таблица **`users`** — информация о пользователях:
- **`user_id`** — уникальный идентификатор пользователя;
- **`first_name`** — имя пользователя;
- **`last_name`** — фамилия пользователя;
- **`age`** — возраст пользователя;
- **`reg_date`** — дата подключения тарифа (день, месяц, год);
- **`churn_date`** — дата прекращения пользования тарифом (если значение пропущено, значит, тариф ещё действовал на момент выгрузки данных);
- **`city`** — город проживания пользователя;
- **`tariff`** — название тарифного плана.
 
Таблица **`calls`** — информация о звонках:
- **`id`** — уникальный номер звонка;
- **`call_date`** — дата звонка;
- **`duration`** — длительность звонка в минутах;
- **`user_id`** — идентификатор пользователя, сделавшего звонок.

Таблица **`messages`** — информация о сообщениях:
- **`id`** — уникальный номер сообщения;
- **`message_date`** — дата сообщения;
- **`user_id`** — идентификатор пользователя, отправившего сообщение.

Таблица **`internet`** — информация об интернет-сессиях:
- **`id`** — уникальный номер сессии;
- **`mb_used`** — объём потраченного за сессию интернет-трафика (в мегабайтах);
- **`session_date`** — дата интернет-сессии;
- **`user_id`** — идентификатор пользователя.

Таблица **`tariffs`** — информация о тарифах:
- **`tariff_name`** — название тарифа;
- **`rub_monthly_fee`** — ежемесячная абонентская плата в рублях;
- **`minutes_included`** — количество минут разговора в месяц, включённых в абонентскую плату;
- **`messages_included`** — количество сообщений в месяц, включённых в абонентскую плату;
- **`mb_per_month_included`** — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах);
- **`rub_per_minute`** — стоимость минуты разговора сверх тарифного пакета;
- **`rub_per_message`** — стоимость отправки сообщения сверх тарифного пакета;
- **`rub_per_gb`** — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета.

## Подготовка данных

Приведем к типу *datetime* столбцы:
- **'reg_date'** и **'churn_date'** из таблицы **`users`**;
- **'call_date'** из таблицы **`calls`**;

```python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y-%m-%d')
users['churn_date'] = pd.to_datetime(users['churn_date'], format='%Y-%m-%d')
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y-%m-%d')
messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y-%m-%d')
sessions['session_date'] = pd.to_datetime(sessions['session_date'], format='%Y-%m-%d')
```

Округлим значения столбца **'duration'** из таблицы **`calls`** вверх и приведем столбец к типу *int*:

```python
calls['duration']  = np.ceil(calls['duration']).astype(int)
```

Создадим столбец **'month'** в таблицах **`calls`**, **`messages`** и **`sessions`** с номером месяца из столбцов **'call_date'**, **'message_date'** и **'session_date'** соответственно:

```python
calls['month'] = calls['call_date'].dt.month
messages['month'] = messages['message_date'].dt.month
sessions['month'] = sessions['session_date'].dt.month
```

Теперь посчитаем для каждого пользователя по месяцам:
- количество сделанных звонков;
- количество израсходованных минут разговора;
- количество отправленных сообщений;
- количество потраченных мегабайт интернет-трафика:

Количество сделанных звонков:

```python
calls_per_month = calls.groupby(['user_id', 'month']).agg(calls=('duration', 'count'))
```

```python
calls_per_month.head(20)
```

Количество израсходованных минут:

```python
minutes_per_month = calls.groupby(['user_id', 'month']).agg(minutes=('duration', 'sum'))
```

```python
minutes_per_month.head(20)
```

Количество отправленных сообщений:

```python
messages_per_month = messages.groupby(['user_id', 'month']).agg(messages=('message_date', 'count'))
```

```python
sessions_per_month = sessions.groupby(['user_id', 'month']).agg({'session_date': 'count'})
```

```python
sessions_per_month.head(20)
```

Количество потраченных мегабайт интернет-трафика:

```python
mb_used_per_month = sessions.groupby(['user_id', 'month']).agg({'mb_used': 'sum'})
```

Округлим в большую сторону до Мб по условиям компании и приведем к целочисленному типу:

```python
mb_used_per_month.head(20).to_frame()
```

Теперь объединим все посчитанные значения в один датафрейм **`user_behavior`**.

Для каждой пары "пользователь-месяц" будут доступны информация о тарифе, количестве звонков, сообщений, потраченных минут и мегабайт трафика.

```python
user_behavior = calls_per_month\
    .merge(minutes_per_month, left_index=True, right_index=True, how='outer')\
    .merge(messages_per_month, left_index=True, right_index=True, how='outer')\
    .merge(sessions_per_month, left_index=True, right_index=True, how='outer')\
    .merge(mb_used_per_month, left_index=True, right_index=True, how='outer')\
    .reset_index()\
    .merge(users, how='left', left_on='user_id', right_on='user_id')
```

Вероятно, некоторые из пользователей за рассматриваемый период времени ни разу не звонили, или ни разу не отправляли сообщения, или ни разу не выходили в интернет. Таких пользователей нет, значит, все пользователи активные, просто некоторые в какие-то периоды не пользовались одним из типов связи. Поэтому заполним образовавшиеся пропуски в данных нулями, которые будут означать, что в конкретном месяце данный пользователь не звонил / не отправлял сообщения / не пользовался интернетом:

```python
user_behavior['calls'] = user_behavior['calls'].fillna(0)
user_behavior['minutes'] = user_behavior['minutes'].fillna(0)
user_behavior['messages'] = user_behavior['messages'].fillna(0)
user_behavior['sessions'] = user_behavior['sessions'].fillna(0)
user_behavior['mb_used'] = user_behavior['mb_used'].fillna(0)
```

```python
user_behavior['calls'] = user_behavior['calls'].astype(int)
user_behavior['minutes'] = user_behavior['minutes'].astype(int)
user_behavior['messages'] = user_behavior['messages'].astype(int)
user_behavior['sessions'] = user_behavior['sessions'].astype(int)
user_behavior['mb_used'] = user_behavior['mb_used'].astype(int)
```

Посчитаем количество минут разговора, сообщений и мегабайт, превышающих тарифный лимит:

```python
user_behavior['paid_minutes'] = user_behavior['minutes'] - user_behavior['minutes_included']
user_behavior['paid_messages'] = user_behavior['messages'] - user_behavior['messages_included']
user_behavior['paid_mb'] = user_behavior['mb_used'] - user_behavior['mb_per_month_included']

for col in ['paid_messages', 'paid_minutes', 'paid_mb']:
    user_behavior.loc[user_behavior[col] < 0, col] = 0
```

Переведем превышающие тарифный лимит мегабайты в гигабайты и сохраним в столбец **'paid_gb'**:

```python
user_behavior['paid_gb'] = np.ceil(user_behavior['paid_mb'] / 1024).astype(int)
```

Посчитаем выручку за дополнительные минуты разговора, сообщения и интернет:

```python
user_behavior['cost_minutes'] = user_behavior['paid_minutes'] * user_behavior['rub_per_minute']
user_behavior['cost_messages'] = user_behavior['paid_messages'] * user_behavior['rub_per_message']
user_behavior['cost_gb'] = user_behavior['paid_gb'] * user_behavior['rub_per_gb']
```

```python
user_behavior['total_cost'] = user_behavior[['rub_monthly_fee', 'total_overpayment']].sum(axis=1)
```

### География пользователей

Составим географическую карту присутствия клиентов с помощью ***Folium***.

```python
# Сохраняем в отдельный датафрэйм координаты городов пользователей
users_coords = users[['latitude', 'longitude']].copy()
```

***Folium*** имеет особенность: при прорисовке *HeatMap* не учитывает вес (количество повторений) каждой точки, и на карте точки с различным количеством повторений имеют одинаковую интенсивность.

Добавим к координатам каждой точки небольшие рандомные числа (несколько сотых), чтобы каждая точка прорисовывалась отдельно в пределах своего города. Для этого создадим функцию:

```python
def rand_coord(row):
    rand_number_1 = random.uniform(-1,1) / 10
    rand_number_2 = random.uniform(-1,1) / 10
    return row['latitude'] + rand_number_1, row['longitude'] + rand_number_2
```

```python
for i in range(len(users_coords)):
    users_coords.loc[i] = rand_coord(users_coords.loc[i])
```

Рисуем карту:

```python
# Создаем базовую карту и ориентируем ее относительно географического центра России
users_map = folium.Map(location=[66.25, 94.15], control_scale=False, zoom_start=3)

# Наносим точки на карту
HeatMap(data=users_coords, radius=10, max_zoom=50, gradient={
    .4: 'blue', .6: 'green', .8: 'yellow', 1.0: 'red'}, blur=7).add_to(users_map)

users_map
```

На карте хорошо видно, что большинство пользователей сконцентрировано в европейской части России.

## Статистический анализ

### Проверка гипотезы о различии средней выручки с пользователей тарифов «Смарт» и «Ультра»

Сформулируем нулевую и альтернативную гипотезы. За нулевую гипотезу традиционно возьмем гипотезу о равенстве:

- **Н₀ (нулевая гипотеза):** средние выручки с пользователей тарифов «Смарт» и «Ультра» равны.
- **Н₁ (альтернативная гипотеза):** средние выручки с пользователей тарифов «Смарт» и «Ультра» отличаются.

Наша нулевая гипотеза — гипотеза о равенстве средних двух генеральных совокупностей. Применим двусторонний тест, так как нам сейчас неважно, у какого тарифа больше средняя выручка, если они отличаются:

```python
alpha = .05
```

```python
smart = user_behavior.loc[user_behavior['tariff'] =='smart']['total_cost']
ultra = user_behavior.loc[user_behavior['tariff'] =='ultra']['total_cost']
```

```python
results = st.ttest_ind(smart, ultra, equal_var = False)

print('p-value:', '{:.1}'.format(results.pvalue))

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Таким образом на имеющихся данных на уровне статистической значимости 5% есть основания отвергнуть нулевую гипотезу о равенстве средних значений выручки с пользователей тарифов «Смарт» и «Ультра» в пользу альтернативной гипотезы. Значит, наша изначальная гипотеза верна: средние показатели выручки двух тарифов отличаются. Проверим, в чью пользу. Сформулируем новую альтернативную гипотезу:

- **Н₀ (нулевая гипотеза):** средние выручки с пользователей тарифов «Смарт» и «Ультра» равны.

- **Н₁ₐ (новая альтернативная гипотеза):** средняя выручка с пользователей тарифа «Ультра» больше, чем с пользователей тарифа «Смарт».

Применим односторонний тест для тарифа «Ультра»:

```python
print('p-value:', '{:.1}'.format(results.pvalue / 2))

if (results.pvalue / 2 < alpha) and (ultra.mean() > smart.mean()):
    print('Отвергаем нулевую гипотезу: \
cредняя выручка с пользователей тарифа «Ультра» больше, чем с пользователей тарифа «Смарт»')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Проверим вручную:

```python
print('Средняя выручка с пользователей тарифа «Смарт»:', '{:.0f} рублей'.format(smart.mean()))
print('Средняя выручка с пользователей тарифа «Ультра»:', '{:.0f} рублей'.format(ultra.mean()))
```

Наша новая альтернативная гипотеза оказалась верной: cредняя выручка с пользователей тарифа «Ультра» оказалась больше, чем с пользователей тарифа «Смарт».

### Проверка гипотезы о различии средней выручки с пользователей из Москвы и из других городов

Сформулируем нулевую и альтернативную гипотезы:

- **Н₀ (нулевая гипотеза):** средние выручки с пользователей из Москвы и из других городов равны.

- **Н₁ (альтернативная гипотеза):** средние выручки с пользователей из Москвы и из других городов отличаются.

Нулевая гипотеза — гипотеза о равенстве средних двух генеральных совокупностей. Применим двусторонний тест:

```python
moscow = user_behavior.loc[user_behavior['city'] =='Москва']['total_cost']
other_cities = user_behavior.loc[user_behavior['city'] !='Москва']['total_cost']
```

```python
results = st.ttest_ind(moscow, other_cities, equal_var = False)

print('p-value:', '{:.2f}'.format(results.pvalue / 2))

if (results.pvalue / 2 < alpha) and (ultra.mean() > smart.mean()):
    print('Отвергаем нулевую гипотезу: \
cредняя выручка с пользователей из Москвы больше, чем с пользователей из других городов')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Видим, что на имеющихся данных на уровне статистической значимости 5% отвергнуть нашу нулевую гипотезу не удалось. Значит, наша изначальная гипотеза статистически не подтвердилась.

Проверим вручную:

```python
print('Средняя выручка с пользователей из Москвы:', '{:.0f} рублей'.format(moscow.mean()))
print('Средняя выручка с пользователей из других городов:', '{:.0f} рублей'.format(other_cities.mean()))
```

На самом деле средние размеры выручки с пользователей из Москвы и из других городов отличаются, но не слишком значительно.
