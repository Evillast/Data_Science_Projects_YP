### Технологический процесс

Золото из руды получают следующим образом.

Когда добытая руда проходит первичную обработку, получается дроблёная смесь. Её отправляют на флотацию (обогащение) и двухэтапную очистку.

**1. Флотация**

Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).

**2. Очистка**

Черновой концентрат проходит две стадии очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

### Структура данных

Данные находятся в трёх таблицах: **обучающая выборка**, **тестовая выборка** и **исходные данные**.

Данные индексируются датой и временем получения информации (признак **date**). Соседние по времени параметры часто похожи.

В тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков. Исходный датасет содержит обучающую и тестовую выборки со всеми признаками.

**Технологический процесс:**

- **`Rougher feed`** — исходное сырье;
- **`Rougher additions`** (или **`reagent additions`**) — флотационные реагенты: **Xanthate, Sulphate, Depressant**;
   - **`Xanthate`** — ксантогенат (промоутер, или активатор флотации);
   - **`Sulphate`** — сульфат (на данном производстве сульфид натрия);
   - **`Depressant`** — депрессант (силикат натрия);
- **`Rougher process`** (англ. *«грубый процесс»*) — флотация;
- **`Rougher tails`** — отвальные хвосты;
- **`Float banks`** — флотационная установка;
- **`Cleaner process`**  — очистка;
- **`Rougher Au`** — черновой концентрат золота;
- **`Final Au`** — финальный концентрат золота.

**Параметры этапов:**

- **`air amount`**  — объём воздуха;
- **`fluid levels`** — уровень жидкости;
- **`feed size`** — размер гранул сырья;
- **`feed rate`** — скорость подачи.

**Наименование признаков**

*[этап].[тип_параметра].[название_параметра]*

**Возможные значения для блока** *[этап]:*

- **`rougher`** — флотация;
- **`primary_cleaner`** — первичная очистка;
- **`secondary_cleaner`** — вторичная очистка;
- **`final`** — финальные характеристики.

**Возможные значения для блока** *[тип_параметра]:*
- **`input `** — параметры сырья;
- **`output`** — параметры продукта;
- **`state`** — параметры, характеризующие текущее состояние этапа;
- **`calculation`** — расчётные характеристики.

А тестовая выборка содержит в себе столбцы, которые являются входными данными, доступными до начала процесса обогащения. Именно с помощью этих данных нужно обучить модель, которая будет предсказывать коэффициент восстановления золота.

Напоследок проверим, правильно ли рассчитана эффективность обогащения руды. Вычислим её на обучающей выборке для признака **rougher.output.recovery** и найдем **MAE** *(среднее абсолютное отклонение)* между нашими расчётами и значением признака.

Эффективность обогащения руды рассчитывается по формуле:

$$\mbox{Recovery} = \frac{C \times (F - T) }{F \times (C - T)} \times 100\%$$

где:
- $C$ — доля золота в концентрате после флотации/очистки;
- $F$ — доля золота в сырье/концентрате до флотации/очистки;
- $T$ — доля золота в отвальных хвостах после флотации/очистки.

Напишем функцию для расчета эффективности обогащения:

```python
def out_recovery(C, F, T):
    recovery = (C * (F - T)) / (F * (C - T)) * 100
    return recovery
```

Сохраним в отдельную таблицу нужные столбцы:

```python
data_recovery = train[['rougher.output.concentrate_au', 'rougher.input.feed_au',
                       'rougher.output.tail_au', 'rougher.output.recovery']].copy()
```

Рассчитаем эффективность обогащения с помощью нашей функции:

```python
data_recovery['rougher.output.recovery_new'] = data_recovery.apply(
    lambda row: out_recovery(row['rougher.output.concentrate_au'],
                             row['rougher.input.feed_au'], 
                             row['rougher.output.tail_au']), axis=1)
```

Посчитаем **MAE**. Значение *среднего абсолютного отклонения* укажет на то, насколько сильно наши расчеты в среднем отличаются от расчетов в таблице.

```python
mae = mean_absolute_error(data_recovery['rougher.output.recovery'],
                          data_recovery['rougher.output.recovery_new'])
print('MAE = {} ≈ {:.2f}'.format(mae, mae))
```

## Разделение признаков

Разобьем выборки на **признаки** и **целевые признаки**.

Первым делом сохраним отдельно тестовые целевые признаки. Сделать это можно, установив соответствие строк в тестовой выборке и исходных данных по значению столбца **'date'**.

Переведем признаки и целевые признаки из формата *float64* в формат *float32*, чтобы уменьшить затраты памяти и увеличить скорость выполнения кода:

```python
features_train = features_train[features_train.columns].astype('float32')
features_test = features_test[features_test.columns].astype('float32')
target_train = target_train[target_train.columns].astype('float32')
target_test = target_test[target_test.columns].astype('float32')
```

Проверим, нет ли в признаках мультиколлинеарности. Выведем на экран список пар столбцов с высокой корреляцией. Чтобы вычислить общую корреляцию по обучающей и тестовой выборкам, будем использовать объединенные признаки.

```python
features_full = pd.concat([features_train, features_test])
```

```python
corr_features_full = np.abs(features_full.corr()).unstack()
multicorr_features_full = corr_features_full[lambda x: ((x >= 0.7) & (x < 1))]
```

```python
multicorr_features_full
```

Видим, что многие столбцы коррелируют со многими другими.

Чтобы понять, сколько столбцов реально имеет смысл удалять (если вообще имеет смысл), чтобы устранить мультиколлинеарность для обучения линейной регрессии, воспользуемся трюком. Составим словарь из частей мультииндекса, соответствующих названию одного коррелирующего признака, состоящий из полностью уникальных ключей и полностью уникальных значений (не встречающихся среди ключей словаря в том числе):

```python
multicorr_to_drop = {}

for ind in multicorr_features_full.sort_values(ascending=False).index:
    if ind[0] not in multicorr_to_drop:
        if ind[0] not in multicorr_to_drop.values():
            if ind[1] not in multicorr_to_drop:
                if ind[1] not in multicorr_to_drop.values():
                    multicorr_to_drop.update({ind[0]: ind[1]})
```

```python
display(multicorr_to_drop)
print()
print(f'Количество столбцов для удаления: {(len(multicorr_to_drop))}')
```

Нам остается только удалить признаки, соответствующие либо ключам, либо значениям словаря. Сохраним оба варианта, чтобы опробовать при дальнейшем обучении модели:

```python
features_train_cut_1 = features_train.drop(features_train[multicorr_to_drop], axis=1)
features_test_cut_1 = features_test.drop(features_train[multicorr_to_drop], axis=1)

features_train_cut_2 = features_train.drop(features_train[multicorr_to_drop.values()], axis=1)
features_test_cut_2 = features_test.drop(features_train[multicorr_to_drop.values()], axis=1)
```

Убедимся на примере обучающей выборки, что все сработало корректно. Датафреймы **features_train_cut_1** и **features_train_cut_2** должны иметь равное количество столбцов и отличаться между собой на 16 столбцов по названию.

```python
find_diff(features_train_cut_1, features_train_cut_2)
```

## Исследование моделей

### Функция для вычисления итогового sMAPE

**sMAPE** *(Symmetric Mean Absolute Percentage Error) — симметричное среднее абсолютное процентное отклонение*.

В отличие от **MAE** выражается не в абсолютных величинах, а в относительных.

Метрика одинаково учитывает масштаб и целевого признака, и предсказания.

Вычисляется по следующей формуле:

$$\mbox{sMAPE} = \frac{1}{N}\sum\limits_{i=1}^N \frac{|y_i - \widehat{y}_i|}{ (|y_i| + |\widehat{y}_i|)/2} \times 100\%$$

где:

- $y_i$ — значение целевого признака для объекта с порядковым номером $i$ в выборке, на которой измеряется качество;
- $\widehat{y}_i$ — значение предсказания для объекта с порядковым номером $i$ в выборке, на которой измеряется качество;
- $N$ — количество объектов в выборке;
- $\sum\limits_{i=1}^N$ — сумма всех объектов выборки ($i$ изменяется от $1$ до $N$).

Нам нужно спрогнозировать сразу две величины:
1. эффективность обогащения чернового концентрата **rougher.output.recovery**;
2. эффективность обогащения финального концентрата **final.output.recovery**.

Таким образом наша итоговая метрика складывается из двух величин:

$$\mbox{Итоговое sMAPE} = 25\% \times \mbox{sMAPE(rougher)} + 75\%\times \mbox{sMAPE(final)}$$

Напишем функцию:

```python
def final_smape(target, pred):
    target = target.to_numpy()
    smape_rougher = (np.divide(2 * (np.abs(target[:, 0] - pred[:, 0])),
                               np.abs(target[:, 0]) + np.abs(pred[:, 0]))).mean() * 100
    smape_final = (np.divide(2 * (np.abs(target[:, 1] - pred[:, 1])),
                             np.abs(target[:, 1]) + np.abs(pred[:, 1]))).mean() * 100
    return 0.25*smape_rougher + 0.75*smape_final
```

Превратим функцию в счетчик для кросс-валидации:

```python
scorer = make_scorer(final_smape, greater_is_better=False)
```

### Линейная регрессия

**Линейная регрессия** — алгоритм машинного обучения, в котором зависимость переменной $x$ от других переменных выражается линейной функцией зависимости.

Исследуем модель с помощью кросс-валидации на обучающей выборке. Хотя масштабирование признаков не влияет на результаты линейной регрессии, оно сделает ее более интерепретируемой. Поэтому стандартизируем признаки с помощью **StandardScaler** и **Pipeline**:

```python
def lin_reg(features):
    model = LinearRegression()
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('model', model)])
    scores = cross_val_score(pipe, features, target_train, scoring=scorer, cv=5)
    result = scores.mean()
    print('Итоговое sMAPE модели LinearRegression: {:.2f}%'.format(np.abs(result)))
```

### Регрессия Лассо

**Lasso** *(Least absolute shrinkage and selection operator)* — алгоритм машинного обучения, использующий метод оценивания коэффициентов линейной регрессионной модели. Метод заключается во введении ограничения на норму вектора коэффициентов модели, что приводит к обращению некоторых коэффициентов в 0 *(L1-регуляризация)*. Метод приводит к повышению устойчивости модели и позволяет получить интерпретируемые модели, — отбираются признаки, оказывающие наибольшее влияние на вектор ответов.

Для подбора оптимальных гиперпараметров модели используем **GridSearchCV**. Здесь и далее будем использовать функции, которые возвращают модели с лучшими гиперпараметрами.

Для стандартизации признаков при кросс-валидации линейных моделей с регуляризацией также используем **StandardScaler** и **Pipeline**.

```python
def lasso(features):
    model = Lasso()
    param_grid = {'model__alpha': [0.3, 0.4, 0.5, 0.6, 0.7]}
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('model', model)])
    search = GridSearchCV(pipe, param_grid, cv=5, scoring=scorer, n_jobs=-1)
    search.fit(features, target_train)
    result = search.best_score_
    best_params = search.best_params_
    best_model = search.best_estimator_
    print('Итоговое sMAPE модели Lasso: {:.2f}%\nС лучшими гиперпараметрами: {}'.format(
        np.abs(result), best_params))
    return best_model
```

```python
best_lasso = lasso(features_train)
```

Модель работает значительно лучше, чем линейная регрессия.

### Мультизадачное Лассо

**Мультизадачное Лассо** — линейная модель, обученная смешанным методом ( *L1/L2 регуляризация*), которая совместно оценивает разреженные коэффициенты для задач множественной регрессии. При *L1-регуляризации* дополнительный член функции потерь является модулем, при *L2-регуляризации* — квадратичной функцией.

```python
def multi_lasso(features):
    model = MultiTaskLasso()
    param_grid = {'model__alpha': [0.4, 0.5, 0.6, 0.7, 0.8]}
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('model', model)])
    search = GridSearchCV(pipe, param_grid, cv=5, scoring=scorer, n_jobs=-1)
    search.fit(features, target_train)
    result = search.best_score_
    best_params = search.best_params_
    best_model = search.best_estimator_
    print('Итоговое sMAPE модели MultiTaskLasso: {:.2f}%\nС лучшими гиперпараметрами: {}'.format(
        np.abs(result), best_params))
    return best_model
```

```python
best_multi_lasso = multi_lasso(features_train)
```

Результат модели оказался хуже, чем у обычной модели Лассо, которая использует только *L1-регуляризацию*.

### Мультизадачная эластичная сетка

**Мультизадачная эластичная сетка** — еще одна линейная модель, использующая как *L1-*, так и *L2-нормированную регуляризацию* коэффициентов,  которая совместно оценивает разреженные коэффициенты для задач множественной регрессии.

```python
def multi_en(features):
    model = MultiTaskElasticNet()
    param_grid = {'model__alpha': [0.7, 0.8, 0.9],
                  'model__l1_ratio': [0.1, 0.2, 0.3]}
    pipe = Pipeline(steps=[('scaler', StandardScaler()), ('model', model)])
    search = GridSearchCV(pipe, param_grid, cv=5, scoring=scorer, n_jobs=-1)
    search.fit(features, target_train)
    result = search.best_score_
    best_params = search.best_params_
    best_model = search.best_estimator_
    print('Итоговое sMAPE модели MultiTaskElasticNet: {:.2f}%\nС лучшими гиперпараметрами: {}'.format(
        np.abs(result), best_params))
    return best_model
```

```python
best_multi_en = multi_en(features_train)
```

Результат модели оказался даже немного лучше, чем у модели Лассо, однако стоит обратить внимание, что модель показывает лучшие результаты, когда максимизирует коэффициент *L1-регуляризации* (которая как раз используется в модели Лассо) и минимизирует коэффициент *L2-регуляризации*.

### Дерево регрессии

**Дерево решений** — универсальный алгоритм машинного обучения. Структура дерева представляет собой «листья» и «ветки». На рёбрах («ветках») дерева решения записаны признаки, от которых зависит целевая переменная, а в «листьях» записаны значения целевой переменной. Чтобы найти значение целевой переменной, надо спуститься по дереву до листа и выдать соответствующее значение. **Дерево регрессии** генерируется, когда целевая переменная является непрерывной.

```python
def tree(features):
    param_grid = {'max_depth': [3, 4, 5],
                  'min_weight_fraction_leaf': [0.0, 0.1],
                  'max_features': [0.6, 0.7]}
    model = DecisionTreeRegressor(random_state=12345)
    search = GridSearchCV(model, param_grid, cv=5, scoring=scorer, n_jobs=-1)
    search.fit(features, target_train)
    result = search.best_score_
    best_params = search.best_params_
    best_model = search.best_estimator_
    print('Итоговое sMAPE модели DecisionTreeRegressor: {:.2f}%\nС лучшими гиперпараметрами: {}'.format(
        np.abs(result), best_params))
    return best_model
```

```python
best_tree = tree(features_train)
```

С помощью точного подбора гиперпараметров мы получили вполне неплохой результат для одиночного дерева.

### Случайный лес

**Случайный лес** —  алгоритм машинного обучения, использующий большой ансамбль независимых друг от друга решающих деревьев. В задаче регрессии их ответы усредняются.

```python
def rand_forest(features):
    param_grid = {'n_estimators': [65, 70, 75],
                  'max_depth': [4, 5, 6],
                  'min_weight_fraction_leaf': [0.0, 0.1],
                  'max_features': [0.6, 0.7]}
    model = RandomForestRegressor(random_state=12345)
    search = GridSearchCV(model, param_grid, cv=5, scoring=scorer, n_jobs=-1)
    search.fit(features, target_train)
    result = search.best_score_
    best_params = search.best_params_
    best_model = search.best_estimator_
    print('Итоговое sMAPE модели RandomForestRegressor: {:.2f}%\nС лучшими гиперпараметрами: {}'.format(
        np.abs(result), best_params))
    return best_model
```

```python
best_forest = rand_forest(features_train)
```

Пока что модель случайного леса продемонстрировала лучший результат среди всех моделей.
