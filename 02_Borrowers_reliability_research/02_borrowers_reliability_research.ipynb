## Предобработка данных

### Обработка пропусков

Выведем на экран количество пропусков в столбцах и попробуем разобраться в причинах их появления:

```python
pd.DataFrame(data.isna().sum(), columns=['NaN']).style.background_gradient('coolwarm')
```

Первое, на что обращаем внимание — количество пропусков в столбцах **'days_employed'** и **'total_income'** совпадает. Проверим: возможно, это одни и те же строки.

```python
print('Строк с парными пропусками:',
      data.query('days_employed.isna() and total_income.isna()').shape[0])
```

Наше предположение оказалось верным: все пропуски в обоих столбцах встречаются в одних и тех же строках.

Теперь проверим, не зависят ли пропуски от типа занятости. Выведем на экран все типы занятости, которые есть в датафрэйме: 

```python
data['income_type'].unique().tolist()
```

Возможно, значения пропущены у безработных граждан или студентов. Проверим:

```python
data.query('days_employed.isna() and total_income.isna()')['income_type'].value_counts().to_frame()
```

Складывается впечатление, что пропуски не зависят от типа занятости заемщика — столбца **'income_type'**.

Можно было бы предположить, что некоторые заемщики намеренно не указали свой трудовой стаж и уровень дохода, если банк допускает такую возможность. Однако маловероятно, что каждый из них не указал одновременно и то, и другое, и среди них (а это 2174 человека) не было ни одного заемщика, который не указал что-то одно, но указал другое. 
Поэтому, скорее всего, эти пропуски неслучайные — возникла некая техническая ошибка при формировании датасета.

На сумму дохода сильнее всего влияет тип занятости, поэтому логично будет заполнить пропуски в столбце **'total_income'** медианным значением по каждому типу занятости из столбца **'income_type'**. Пропуски в столбце **'days_employed'** можно заполнить медианным значением по типу занятости и возрасту из столбца **'dob_years'**.

Однако, прежде чем заполнять пропуски, стоит обработать аномальные значения. По крайней мере в столбце **'days_employed'** их настолько много, что они могут повлиять даже на медиану, несмотря на то, что медиана лучше, чем среднее, описывает наиболее веротяные значения для выборки, когда в данных имеются артефакты.

### Обработка аномальных значений

Начнем со столбца **'dob_years'**. В нем встречаются аномальные строки с нулевым возрастом. Так как эта информация важна для дальнейшего исследования, а восстановить ее сложно, удалим эти строки:

```python
data = data.query('dob_years != 0')
```

Проверим:

```python
print(f'Минимальный возраст заемщиков: {data["dob_years"].min()} лет')
```

Теперь изучим численный состав категорий занятости в столбце **'income_type'**:

```python
data['income_type'].value_counts().to_frame()
```

У нас присутствуют 4 крайне млочисленные категории занятости: «безработный», «предприниматель», «студент» и «в декрете». Поступим с ними следующим образом: категорию «предприниматель» объединим с категорией «компаньон», категорию «в декрете» с категорией «сотрудник» (чтобы уйти в декрет, нужно работать по найму, либо состоять на гос. службе, но это статистически менее вероятно), аномальные категории «безработный» и «студент» просто удалим, так как таким категориям заемщиков обычно в принципе не выдают кредитов.

```python
data['income_type'] = data['income_type'].replace('предприниматель', 'компаньон')
data['income_type'] = data['income_type'].replace('в декрете', 'сотрудник')
```

```python
data = data.query('income_type != "безработный" and income_type != "студент"')
```

Проверим:

```python
data['income_type'].value_counts().to_frame()
```

Теперь разберемся со столбцом **'days_employed'**. В нем встречаются странные отрицательные значения трудового стажа. Заменим их на положительные:

```python
data['days_employed'] = data['days_employed'].abs()
```

Для каждого типа занятости выведем медианное и минимальное значения трудового стажа из столбца **'days_employed'**:

```python
data.groupby('income_type')['days_employed'].agg(['median', 'min'])
```

У пенсионеров отображаются аномально большие значения. Нужно с ними что-то сделать.

Необходимые для выхода на пенсию трудовой стаж и возраст претерпевают изменения, кроме того, зависят от сферы деятельности. Проверим минимальный возраст пенсионеров в нашем датафрэйме: 

```python
print(f'Минимальный возраст заемщиков-пенсионеров: {data[data["income_type"] == "пенсионер"]["dob_years"].min()} года')
```

Снова обнаружили странное значение. Выведем количество строк, в которых пенсионерам меньше 50 лет:

```python
print('Количество заемщиков-пенсионеров младше 50 лет:',
      data.query('income_type == "пенсионер" and dob_years < 50')['dob_years'].count())
```

Таких строк довольно много. Возможно, это люди с ограниченными возможностями здоровья и представители профессий, где предусмотрен досрочный выход на пенсию.

Поступим следующим образом. Для пенсионеров возрастом от 53 лет заменим трудовой стаж на условные 35 лет (12775 дней). Для пенсионеров от 38 до 52 лет заменим трудовой стаж на разность возраста и 18 лет в днях. Более молодых пенсионеров будем считать нерабочими категориями граждан и заменим их трудовой стаж на 0.

```python
data.loc[(data['income_type'] == "пенсионер")
         & (data['dob_years'] >= 53), 'days_employed'] = 12775
data.loc[(data['income_type'] == "пенсионер")
         & (data['dob_years'] >= 38)
         & (data['dob_years'] < 53), 'days_employed'] = (data['dob_years'] - 18) * 365
data.loc[(data['income_type'] == "пенсионер")
         & (data['dob_years'] < 38), 'days_employed'] = 0
```

Проверим:

```python
data.groupby('income_type')['days_employed'].agg(['median', 'min'])
```

Теперь выглядит приемлемо.

Разберемся с артефактами в столбце **'children'**. Выведем количество уникальных значений в столбце:

```python
data['children'].value_counts().to_frame()
```

Там, где стоит значение -1, возможно, должно стоять 1 или 0, а там, где стоит 20, должно стоять 2 либо некое относительно большое число (больше 5). Удалим эти строки:

```python
data = data.query('children != -1 and children != 20')
```

Проверим, что артефакты удалены:

```python
data['children'].unique().tolist()
```

Теперь проверим, насколько много в датафрэйме заемщиков с аномально большими доходами (более 700 тысяч рублей, судя по гистограмме):

Таких строк обнаружилось довольно много. Верится с трудом, но оставим их как есть, поскольку невозможно однозначно установить недостоверность информации.

### Обработка пропусков (продолжение)

Заполним пропуски в столбце **'total_income'** медианным значением по каждому типу занятости из столбца **'income_type'**:

```python
data['total_income'] = data['total_income'].fillna(
    data.groupby('income_type')['total_income'].transform(lambda x: x.median()))
```

Пропуски в столбце **'days_employed'** заполним медианным значением по типу занятости и возрасту:

```python
data['days_employed'] = data['days_employed'].fillna(
    data.groupby(['income_type', 'dob_years'])['days_employed'].transform(lambda x: x.median()))
```

Проверим, остались ли пропуски:

### Категоризация данных

#### Лемматизация и стемминг

Для того, чтобы категоризировать данные в столбце **'purpose'**, нам нужно составить список уникальных *лемм* и найти их основы.

Для начала выведем на экран перечень уникальных целей получения кредита из столбца **'purpose'**:

```python
data['purpose'].value_counts().to_frame()
```

*Лемматизируем* данные цели с помощью модуля **pymystem3**:

```python
m = Mystem()

lemmas = []
for i in tqdm(range(len(data['purpose'].value_counts().index))):
    lemma = m.lemmatize(data['purpose'].value_counts().index[i])
    lemmas.extend(lemma * data['purpose'].value_counts()[i])
```

Составим рейтинг уникальных лемм:

```python
unique_lemmas = Counter(lemmas)
sorted(unique_lemmas.items(), key = lambda x: x[1], reverse=True)
```

Возьмем топ-7 слов, исключив пробелы и переносы строк, а также слова длиной менее 5 букв (предлоги, местоимения), чтобы получить список существительных. Дальше топ-7 идут либо существительные со схожим смыслом, либо определения и глаголы, которые нам не понадобятся для категоризации.

Поместим топ-7 слов в отдельный список:

```python
top_purposes = [lemma for lemma in sorted(unique_lemmas, key=unique_lemmas.get, reverse=True)
                if len(lemma) > 4][:7]
print(top_purposes)
```

Сократим полученный список до топ-5. Уберем из списка общие понятия «покупка» и «операция»:

```python
top_purposes.remove('покупка') 
top_purposes.remove('операция')
print(top_purposes)
```

Для того, чтобы произвести категоризацию, сначала проведем *стемминг* — найдем основы слов из данного списка с помощью **nltk.stem**:

```python
stemmer = SnowballStemmer('russian')

stemmed_purposes = [stemmer.stem(word) for word in top_purposes]
print(stemmed_purposes)
```

Теперь можем приступать к категоризаии данных.

#### Категоризация

Разобьем заемщиков на категории:

- **по возрасту**:
  - `до 30 лет`;
  - `31—45 лет`;
  - `46—60 лет`;
  - `старше 60 лет`;
  

- **по трудовому стажу**:
  - `до 5 лет`;
  - `6—10 лет`;
  - `11—20 лет`;
  - `больше 20 лет`;
  

- **по уровню дохода**:
  - до 50000 — `D`;
  - 50001–200000 — `C`;
  - 200001–700000 — `B`;
  - от 700001 — `A`;


- **по количеству детей**:
  - `нет детей`;
  - `один ребенок`;
  - `двое детей`;
  - `многодетный`;


- **по цели получения кредита**:
  - `'операции с недвижимостью'`;
  - `'операции с автомобилем'`;
  - `'получение образования'`;
  - `'проведение свадьбы'`.

Создадим класс **Categorizer**, а в качестве методов класса напишем функции для категоризации данных:

```python
class Categorizer:
    
    def сategorize_age(self, age):
        if age <= 30:
            return 'до 30 лет'
        elif 30 < age <= 45:
            return '31—45 лет'
        elif 45 < age <= 60:
            return '46—60 лет'
        return 'старше 60 лет'
        
    def сategorize_seniority(self, seniority):
        if seniority <= 1825:
            return 'до 5 лет'
        elif 1825 < seniority <= 3650:
            return '6—10 лет'
        elif 3650 < seniority <= 7300:
            return '11—20 лет'
        return 'больше 20 лет'
        
    def сategorize_income(self, income):
        if income <= 50000:
            return 'D'
        elif 50000 < income <=200000:
            return 'C'
        elif 200000 < income <=700000:
            return 'B'
        return 'A'
    
    def сategorize_children(self, children):
        if children == 1:
            return 'один ребенок'
        elif children == 2:
            return 'двое детей'
        elif children >= 3:
            return 'многодетный'
        return 'нет детей'
    
    def сategorize_purpose(self, purpose):
        if stemmed_purposes[0] in purpose or stemmed_purposes[1] in purpose:
            return 'операции с недвижимостью'
        elif stemmed_purposes[2] in purpose:
            return 'операции с автомобилем'
        elif stemmed_purposes[3] in purpose:
            return 'получение образования'
        elif stemmed_purposes[4] in purpose:
            return 'проведение свадьбы'
        return 'нет категории'
```

Сформируем новые столбцы **'age_category'**, **'seniority_category'**, **'total_income_category'**, **'children_category'** и **'purpose_category'**, куда поместим категории заемщиков по возрасту, трудовому стажу, уровню дохода, количеству детей и цели получения кредита соответственно:

```python
c = Categorizer()

data['age_category'] = data['dob_years'].apply(c.сategorize_age)
data['seniority_category'] = data['days_employed'].apply(c.сategorize_seniority)
data['total_income_category'] = data['total_income'].apply(c.сategorize_income)
data['children_category'] = data['children'].apply(c.сategorize_children)
data['purpose_category'] = data['purpose'].apply(c.сategorize_purpose)
```

Столбцы **'education'** и **'family_status'** были категоризированы в таблице изначально (столбцы **'education_id'** и **'family_status_id'** соответственно). Проверим, что все корректно:

```python
data[['education', 'education_id']].value_counts().to_frame('count')
```

```python
data['education_id'] = data['education_id'].replace(4, 0)
```

Кроме того, не понятно, по какой логике категориям образования присваивались номера. Чтобы обозначения категорий были интуитивно понятными, заменим категории в столбце **'education_id'** на следующие:

- начальное — `D`;
- среднее — `C`;
- неоконченное высшее	 — `B`;
- высшее/ученая степень — `A`.

```python
data['education_id'] = data['education_id'].replace(3, 'D')
data['education_id'] = data['education_id'].replace(1, 'C')
data['education_id'] = data['education_id'].replace(2, 'B')
data['education_id'] = data['education_id'].replace(0, 'A')
```
